```python
# import Point from Shapely
from shapely.geometry import Point
```

```python
# Initialise coordinate reference system
crs = {'init': 'epsg"4326'}
```

```python
# Create point using latitude & longitude info
#geometry = gpd.points_from_xy(SensorLoc.longitude, Sensor.latitude)
geometry = [Point(xy) for xy in zip(SensorLoc['longitude'], SensorLoc['latitude'])]
geometry
```

```python
# Create a Geo-DataFrame
geo_SensorLoc = gpd.GeoDataFrame(SensorLoc, crs=crs, geometry=geometry)
geo_SensorLoc.head()
```

```python
import folium
```

```python
Melb_CBD = [-37.8124, 144.9623]
m = folium.Map(location = Melb_CBD)

# display the map
display(m)
```

```python
for row in SensorLoc.iterrows():
    row_values=row[1]
    #print(row_values['latitude'])
```

```python
# Allocate sensor location onto map

for row in SensorLoc.iterrows():
    row_values = row[1]
    location = [row_values['latitude'], row_values['longitude']]
    
    marker = folium.CircleMarker(location=location,
                                 color='blue',
                                 fill_color='blue',
                                 fill_opacity=0.5,
                                 radius=5,
                                 tooltip='Sensor ID : '+ str(row_values['sensor_id']))
        
    marker.add_to(m)

display(m)
```
